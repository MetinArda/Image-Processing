### **Part 1:** Cubic spline Low Pass filter

\begin{equation}
y(x) = \Bigg\{
\begin{array}{ll}
1-x^2 & {\rm for}~ |x| \leq 1 \\
-2(x-1)(x-2)^2 & {\rm for}~ 1 < |x| \leq 2 \\
0 & {\rm otherwise}
\end{array}
\end{equation}


```python
# cubic spline LP filter

# calculate cubic spline filter coefficients given filter factor
def my_cubic_filter(n_filter):
    
    assert n_filter >= 1
    
    size_filter = n_filter * 4 - 1
    
    filter_cubic = np.zeros((size_filter,))
    
    filter_cubic[size_filter//2] = 1.0
    
    for i in range(1,n_filter):
        x = i/n_filter
        filter_cubic[size_filter//2 + i] = 1.0 - x**2

    for i in range(n_filter+1, n_filter*2):
        x = i/n_filter
        filter_cubic[size_filter//2 + i] = -2*(x-1)*(x-2)**2
    
    filter_cubic[:size_filter//2] = filter_cubic[-1:size_filter//2:-1]

    return filter_cubic

# end of function
```

```python
# STEP 2 Calculate 1-D cubic spline low-pass filter coefficients

# set effective filter size
n_filter = 2

# calculate filter coefficients
filter_cubic = my_cubic_filter(n_filter)

print("effective filter size: ", n_filter)
print("filter support size: ", filter_cubic.size)
print("1-D filter coefficients: ", filter_cubic)
print("sum of 1-D filter coefficients: ", filter_cubic.sum())

# calculate 2-D filter coefficients (separable filter)
filter_cubic_2D = np.outer(filter_cubic, filter_cubic)

print("2-D filter coefficients: \n", filter_cubic_2D)
print("sum of 2-D filter coefficients: ", filter_cubic_2D.sum())
```

### **Part 2:** Image Decimation


```python
# my function to round image data to nearest integer, then
# truncate to range [0, 255], and then set data type to uint8

def my_imgTruncate(img):
    
    img = np.round(img, 0)
    img = np.minimum(img, 255)
    img = np.maximum(img, 0)
    img = img.astype('uint8')
    
    return img

# end of function
```

```python
# my function to convert to lumincance, round to nearest integer, 
# truncate to range [0, 255], and then set data

def my_imgLuminance(imgRGB):
    
    # make sure it is a color image
    dim_img = imgRGB.shape[2]    
    assert dim_img >= 3
    
    # get the luminance data
    if dim_img == 3:
        imgLum = color.rgb2gray(imgRGB)
    else:
        # ignore the alpha channel
        imgLum = color.rgb2gray(imgRGB[:,:,0:3])

    imgLum = np.round(imgLum * 255, 0)
    imgLum = np.minimum(imgLum, 255)
    imgLum = np.maximum(imgLum, 0)
    imgLum = imgLum.astype('uint8')
    
    return imgLum

# end of function
```

```python
# display an image in original size

def my_display_actual_size(img, str_caption):

    height, width = img.shape

    # determine a figure size big enough to accomodate an axis of xpixels by ypixels
    # as well as the ticklabels, etc.
    margin = 0.05
    dpi = 80
    figsize = (1.0+margin)*height/dpi, (1.0+margin)*width/dpi

    # define the figure
    fig = plt.figure(figsize=figsize, dpi=dpi)
    
    # make the axis the right size
    ax = fig.add_axes([margin, margin, 1 - 2*margin, 1 - 2*margin])

    # display the image
    ax.imshow(img, cmap='gray', vmin=0, vmax=255, interpolation='none')
    plt.title(str_caption)
    plt.show()
    
    return

# end of function
```

```python
# STEP 3.1 Pick an image

# set image folder
image_folder = r'/Users/tanjuerdem/EE421521_2022/images'

# read input image 
image_file = r'/barbara.png'
image_path = image_folder + image_file
imgRGB = io.imread(image_path)

if imgRGB.ndim == 3:
    # calculate the luminance image
    img = my_imgLuminance(imgRGB)
else:
    img = imgRGB

height, width = img.shape
print("Image width is {} and image height is {}.".format(width, height))

# display image in actual size
my_display_actual_size(img,'Input Image')
```

```python
# STEP 3.2 Downsample the image: Remove pixels

# downsampling factor
n_down = 4

# downsample image
img_down = img[::n_down,::n_down]

# display image
my_display_actual_size(img_down, 'Downsampled Image')

# (observe that there is aliasing in parts of the downsampled image)
```

```python
# STEP 3.3 Decimating the image: Remove pixels after low-pass filtering

# Case 1: LP filtering with a mean filter

# downsampling factor
n_down = 4

# set filter coefficients
my_filter = np.ones((n_down, n_down))
my_filter /= my_filter.sum()

# low-pass (LP) filter image
img_dec = signal.convolve2d(img, my_filter, mode='same', boundary='fill', fillvalue=128)

# no low-pass filtering, only subsampling
#img_sub = img

# round to nearest integer
img_dec = my_imgTruncate(img_dec)

# downsample image
img_dec = img_dec[::n_down,::n_down]

# display image
my_display_actual_size(img_dec, 'Decimated Image with Mean LP Filter')

# (observe that there is no aliasing in the decimated image)
```

```python
# STEP 3.3 Decimating the image: Remove pixels after low-pass filtering (cont.)

# Case 2: LP filtering with a cubic spline filter

# downsampling factor
n_down = 4

# set filter coefficients
my_filter = my_cubic_filter(n_down)
my_filter_2D = np.outer(my_filter,my_filter)
my_filter_2D /= my_filter_2D.sum()

# low-pass (LP) filter image
img_dec = signal.convolve2d(img, my_filter_2D, mode='same', boundary='fill', fillvalue=128)

# round to nearest integer
img_dec = my_imgTruncate(img_dec)

# downsample image
img_dec = img_dec[::n_down,::n_down]

# display image
my_display_actual_size(img_dec, 'Decimated Image with Bicubic LP Filter')

# (observe that bicubic LP filter produces a better result than mean filter)
```

---------------------------------------------------------------
### **Part 3:** Image Interpolation



```python
# STEP 4.1 Upsampling the image: Insert zero-valued pixels 

# upsampling factor
n_up = 2

img_up = np.zeros((height*n_up,width*n_up))
   
# upsample image
img_up[::n_up,::n_up] = img

# display image
my_display_actual_size(img_up, 'Upsampled Image')
```

```python
# STEP 4.2 Interpolating the image: Low-pass filter the image after inserting zero-valued pixels 

# Case 1: interpolation via pixel replication

# set filter coefficients
n_up = 2
my_filter = np.ones((n_up, n_up))
# (note: there is no normalization of filter coefficients in the case of interpolation)

# 2D convolution
img_interp = signal.convolve2d(img_up, my_filter, mode='same', boundary='fill', fillvalue=128)

# round to nearest integer
img_interp = my_imgTruncate(img_interp)

# display image
my_display_actual_size(img_interp, 'Interpolated Image via Pixel Replication')
```

```python
# STEP 4.2 Interpolating the image: Low-pass filter the image after inserting zero-valued pixels (cont.)

# Case 2: interpolation with bicubic LP filter

# set filter coefficients
n_up = 2
my_filter = my_cubic_filter(n_up)
my_filter_2D = np.outer(my_filter,my_filter)
# (note: there is no normalization of filter coefficients in the case of interpolation)

# 2D convolution
img_interp = signal.convolve2d(img_up, my_filter_2D, mode='same', boundary='fill', fillvalue=128)

# round to nearest integer
img_interp = my_imgTruncate(img_interp)

# display image
my_display_actual_size(img_interp, 'Interpolated Image with Bicubip LP Filter')

# (observe that bicubic LP filter produces a better result than pixel replication)
```
