```python
# import necessary packages

# reading/writing image files
from skimage import io
from skimage import color

# displaying images and plots
import matplotlib as mpl
import matplotlib.pyplot as plt

# array operations
import numpy as np

# signal processing operations
from scipy import signal

# mathematical calculations
import math

# random number generation
from scipy.stats import norm
```

```python
# calculate the RMSE between two images

def my_RMSE(img1, img2):
    
    assert img1.ndim == img2.ndim
    assert img1.size == img2.size
    
    return math.sqrt(((img1 - img2)**2).sum() / img1.size)
    
# end of function
```

```python
# my function for calculating a Gaussian kernel

def my_winGauss(M):

    assert M > 0
    
    # standard deviation of Gaussian
    sigma = 0.425 * M
    two_sigma_sqr = 2.0 * sigma * sigma
    
    # filter size
    L = int(np.ceil(3.0 * sigma))
    if L%2 == 0:
        L += 1

    win_Gauss = np.zeros(L)
    
    L_half = L//2
    for n in range(L):
        m = n - L_half
        win_Gauss[n] = math.exp(-m*m/two_sigma_sqr)
        
    return win_Gauss
    
```

```python
# my function to convert to lumincance, round to nearest integer, 
# truncate to range [0, 255], and then set data

def my_imgLuminance(imgRGB):
    
    # make sure it is a color image
    dim_img = imgRGB.shape[2]    
    assert dim_img >= 3
    
    # get the luminance data
    if dim_img == 3:
        imgLum = color.rgb2gray(imgRGB)
    else:
        # ignore the alpha channel
        imgLum = color.rgb2gray(imgRGB[:,:,0:3])

    imgLum = np.round(imgLum * 255, 0)
    imgLum = np.minimum(imgLum, 255)
    imgLum = np.maximum(imgLum, 0)
    imgLum = imgLum.astype('uint8')
    
    return imgLum

# end of function
```

```python
# display an image in original size

def my_display_actual_size(img, str_caption):
    
    dpi = mpl.rcParams['figure.dpi']
    height, width = img.shape

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    # ax.axis('off')

    # Display the image.
    ax.imshow(img, cmap='gray', vmin=0, vmax=255, interpolation='none')
    plt.xticks([]), plt.yticks([])

    plt.title(str_caption)
    plt.show()

    return

# end of function
```

```python
# STEP 1 Write a function to implement bilateral filtering

# ADD YOUR CODE HERE

# for every pixel:

def bilateralFilter(noisyImg,W,M):
    
    filteredImg = noisyImg.copy()
    # get image size
    height, width = noisyImg.shape
    
    # for every pixel:
    for i in range(W,height-W):
        for j in range(W,width-W):
            smoothingWeight = 1
            rangeWeight = 1
            # standard deviation of Gaussian
            sigma = 0.425 * M
            two_sigma_sqr = 2.0 * sigma * sigma
            for k in range(i-W,i+W):
                for l in range(j-W,j+W):
                    # calculate smoothing weight (depending on distance in location of pixels)
                    smoothingWeight += math.exp(-((i-k)*2 + (j-l)*2) / (two_sigma_sqr))

                    # calculate range weight (depending on difference in amplitude of pixels)
                    rangeWeight += math.exp(-( (noisyImg[i][j] - noisyImg[k][l] )**2 ) / (two_sigma_sqr))
                                            
            # multiply the above two weights to determine local filter coeffificents
            localFilterCoef = smoothingWeight * rangeWeight
            
            # apply this filter over the local window
            filteredImg[i][j] = noisyImg[i][j] * localFilterCoef
            
    return filteredImg
```

```python
# STEP 2 Add noise to the input image

# ADD YOUR CODE HERE

# choose PSNR value (use 20 dB for this homework)

# get image size
size = imgLum.shape

# Gaussian noise parameters
noise_Mean = 0.0
PSNR = 20.0
noise_Std = 255.0 / (10.0 ** (PSNR/20.0))

# add noise to the input image with the specified PSNR

# add Gaussian noise to image
imgNoisy = imgLum + norm.rvs(noise_Mean, noise_Std, size)

# calculate and print the RMSE value between the original image and the noisy image
rmse_Noisy = my_RMSE(imgLum, imgNoisy)

# display the original image and the noisy image
my_display_actual_size(imgLum, 'Original Image')
my_display_actual_size(imgNoisy,'Noisy Image (RMSE = {:.2f})'.format(rmse_Noisy))
```

```python
# STEP 3 Filter the noisy image with a bileteral filter

# ADD YOUR CODE HERE

# choose a filtering window size (for this homework use 3x3 (M=1.5), 5x5 (M=2.5) and 7x7 (M=4.5) windows)
W1 = 1
W2 = 2
W3 = 3

# use your function to apply the bileteral filter on the noisy image
filteredIMG1 = bilateralFilter(imgNoisy,W1,1.5)
filteredIMG2 = bilateralFilter(imgNoisy,W2,2.5)
filteredIMG3 = bilateralFilter(imgNoisy,W3,4.5)

# calculate and print the RMSE value between the original image and the filtered image
rmse1 = my_RMSE(imgLum, filteredIMG1)
print(rmse1)
rmse2 = my_RMSE(imgLum, filteredIMG2)
print(rmse2)
rmse3 = my_RMSE(imgLum, filteredIMG3)
print(rmse3)

# display original, noisy, and filtered images
my_display_actual_size(imgLum,'Original Image')
my_display_actual_size(imgNoisy,'Noisy Image')
my_display_actual_size(filteredIMG1,'Filtered Image M is 1.5')
my_display_actual_size(filteredIMG2,'Filtered Image M is 2.5')
my_display_actual_size(filteredIMG3,'Filtered Image M is 4.5')
```
