```python
# STEP 0 Import the necessary packages

# use skimage for reading/writing image files
from skimage import io
from skimage import color

# use matplotlib for displaying images and creating plots
import matplotlib.pyplot as plt

# use numpy for array operations
import numpy as np
```

```python
# STEP 1.a Read & display a color image

# set image folder
image_folder = r'C:\Users\altug\OneDrive\Masast\EE421'

# choose an image file
image_file = r'\peppers_more.jpg'

# evaluate full path of image file
image_path = image_folder + image_file

# read color image file
imgRGB = io.imread(image_path)

# display size and data type of image
height = imgRGB.shape[0]
width = imgRGB.shape[1]
band = imgRGB.shape[2]
print("image height =", height)
print("image width =", width)
print("color bands =", band)
print("image data type:", imgRGB.dtype)

# display color image
plt.imshow(imgRGB)
plt.title('Input Color Image')
plt.xticks(np.arange(0,width,25)), plt.yticks(np.arange(0,height,25))
plt.show()
```

```python
# STEP 1.b Read & display a monochrome image

# choose an image file
image_file = r'/peppers_more.jpg'

# evaluate full path of image file
image_path = image_folder + image_file

# read color image file
imgGray = io.imread(image_path)

# display size and data type of image
height = imgGray.shape[0]
width = imgGray.shape[1]
print("image height =", height)
print("image width =", width)
print("image data type:", imgGray.dtype)

# display monochrome image
plt.imshow(imgGray, cmap='gray')
plt.title('Input Monochrome Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# STEP 2.a Print a block of pixels
image_file = r'/cameraman.tiff'
# evaluate full path of image file
image_path = image_folder + image_file

print("A 4x6 block of monochrome image:\n")
print(imgGray[200:204, 120:126])
```

```python
# STEP 2.b Print a block of pixels
image_file = r'/peppers_more.jpg'
# evaluate full path of image file
image_path = image_folder + image_file
print("A 2x4 block of color image:\n")
print(imgRGB[50:52, 60:64])
```

```python
# STEP 2.c Print a row of pixels

row = 85
print("The B values of the first 10 pixels on row {} of color image:\n".format(row))
print(imgRGB[row, :10,])
```

```python
# STEP 2.d Print a column

column = 100
step = 20
print("Every {}th pixel on the {}th column of monochorome image:\n".format(step, column))

print(imgGray[::step, column])
```

```python
# STEP 3.a Display a row-wise slice of a color image

plt.imshow(imgRGB[70:90, :, :])
plt.title('A Row-Wise Slice of Color Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# STEP 3.b Display a block of monochrome image
image_file = r'/cameraman.tiff'
# evaluate full path of image file
image_path = image_folder + image_file
plt.figure(figsize=(2,2))
plt.imshow(imgGray[30:80, 70:130], cmap='gray')
plt.title('A Block of Monochrome Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.close()
```

```python
# STEP 4 Convert color image into gray scale & display it

image_file = r'/peppers_more.jpg'
# evaluate full path of image file
image_path = image_folder + image_file
# convert image into gray scale (find its luminance)
# this function returns number in [0,1] range
imgLum = color.rgb2gray(imgRGB)

# print monochrome image
print(imgLum)

# display monochrome image
plt.imshow(imgLum, cmap='gray', vmin=0, vmax=1) # scales [0,1] to [0,255] during display
plt.title('Monochrome Version of Color Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# STEP 5 Round to nearest integer and set data type to uint8

# multiply with 255 and round to nearest integer
imgLum_int = np.round(imgLum * 255, 0)

# truncate above to 255
imgLum_int = np.minimum(imgLum_int, 255)

# truncate below to 0
imgLum_int = np.maximum(imgLum_int, 0)

# set data type to integer
imgLum_int = imgLum_int.astype(np.uint8)

# print monochrome image
print(imgLum_int)

# display monochrome image
plt.imshow(imgLum_int, cmap='gray')
plt.title('Monochrome Version of Color Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# STEP 6 Write an image into file

# write the monochrome version of the color image into a file
out_file = r'/peppers_more_out.jpg'
out_path = image_folder + out_file
io.imsave(out_path, imgLum_int, quality=100)
print(out_path)
```
