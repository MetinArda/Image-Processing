```python
# STEP 1 Read & display a monochrome image

# set image folder
image_folder = r'C:\Users\altug\OneDrive\Masast\EE421\Week1'

# choose image file
image_file = r'\cameraman.tiff'

# evaluate full path of image file
image_path = image_folder + image_file

# read color image file
imgGray = io.imread(image_path)

# display color image
plt.imshow(imgGray, cmap='gray')
plt.title('Input Color Image')
#plt.xticks([]), plt.yticks([])
plt.show()
```
